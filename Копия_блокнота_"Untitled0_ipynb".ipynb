```python
# advanced_calculator.py

history = []

def calculate():
    operations = {
        '+': lambda x, y: x + y,
        '-': lambda x, y: x - y,
        '*': lambda x, y: x * y,
        '/': lambda x, y: x / y if y != 0 else "Ошибка: деление на ноль",
        '**': lambda x, y: x ** y,  # возведение в степень
        '%': lambda x, y: x % y,     # остаток от деления
    }

    print("Калькулятор с историей")
    print("Доступные операции: +, -, *, /, **, %")
    print("Введите 'history' для просмотра истории, 'clear' для очистки, 'q' для выхода\n")

    while True:
        user_input = input("Введите выражение: ").strip()

        if user_input.lower() == 'q':
            print(" До свидания!")
            break
        elif user_input.lower() == 'history':
            if history:
                print("\n История операций:")
                for i, item in enumerate(history, 1):
                    print(f"{i}. {item}")
            else:
                print(" История пуста")
            continue
        elif user_input.lower() == 'clear':
            history.clear()
            print(" История очищена")
            continue

        try:
            # Разбираем выражение
            for op in ['**', '*', '/', '+', '-', '%']:
                if op in user_input:
                    parts = user_input.split(op)
                    if len(parts) == 2:
                        a, b = float(parts[0]), float(parts[1])
                        result = operations[op](a, b)
                        entry = f"{user_input} = {result}"
                        print(f" {entry}")
                        history.append(entry)
                        break
            else:
                print("❌ Не распознано выражение")
        except ValueError:
            print(" Введите корректные числа")
        except Exception as e:
            print(f" Ошибка: {e}")

if __name__ == "__main__":
    calculate()
```
